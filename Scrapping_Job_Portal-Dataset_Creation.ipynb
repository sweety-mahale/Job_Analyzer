```python
import numpy as np
import pandas as pd
from tqdm import tqdm
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
browser=webdriver.Chrome()
```

```python
data = []

for page in tqdm(range(20,3000)):    

    link = 'https://www.naukri.com/jobs-in-india-' + str(page)
    browser.get(link)
    time.sleep(2)
    soup = BeautifulSoup(browser.page_source, 'html.parser')
    
    for job in soup.find_all('div',class_="srp-jobtuple-wrapper"):    
    
        try:
            job_id          = job.get('data-job-id').strip()
        except:
            job_id          = np.nan
        try:
            job_role        = job.find('a' , class_ = 'title').text.strip()
        except:
            job_role        = np.nan   
        try:    
            job_link        = job.find('a' , class_ = 'title').get('href')
        except:
            job_link        = np.nan  
        try:
            reviews         = job.find('a' , class_ = 'review ver-line').text.strip()
        except:
            reviews         = np.nan   
        try:    
            rating          = job.find('a' , class_ = 'rating').find('span',class_='main-2').text.strip()
        except:
            rating          = np.nan 
        try:
            company         = job.find('a' , class_ = 'comp-name mw-25').text.strip()
        except:
            company         = np.nan    
        try:
            company_link    = job.find('a' , class_ = 'comp-name mw-25').get('href')
        except:
            company_link    = np.nan    
        try:
            experience      = job.find('span',class_='expwdth').text.strip()
        except:
            experience      = np.nan    
        try:
            salary          = job.find('span',class_='sal-wrap ver-line').text.strip()
        except:
            salary          = np.nan   
        try:
            location        = job.find('span',class_='locWdth').text.strip()
        except:
            location        = np.nan  
        try:
            Key_Skills      = ','.join([i.text.strip() for i in job.find('ul',class_ = 'tags-gt').find_all('li')])
        except:
            Key_Skills      = np.nan
        try:
            posted_on       = job.find('span',class_='job-post-day').text.strip()
        except:
            posted_on       = np.nan


        data.append([job_id, job_role, company, experience,salary,location,rating,reviews,
                     Key_Skills, posted_on ,job_link, company_link])
```

```python
df = pd.DataFrame(data, columns = ['job_id', 'job_role', 'company', 'experience','salary','location',
                                   'rating','reviews','resposibilities', 'posted_on' ,'job_link', 
                                   'company_link'])

df.head()
```

```python
df.to_csv('jobs_data.csv', index = False)
```
